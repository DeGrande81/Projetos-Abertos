## Contagem de bicicletas naponte Fremont, em Seatle.

```python
import pandas as pd
data = pd.read_csv('fremont-bridge.csv', index_col = 'Date', parse_dates = True)
data.head()
```

```python
data.columns = ['West', 'East']
data ['Total'] = data ['West'] + data ['East']
data.head()
```

```python
data.dropna().describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
data.plot()
plt.ylabel('Contagem de Bicicletas por hora')
plt.show()
```

```python
weekly = data.resample('W').sum()
weekly.plot(style = [':', '--', '-'])
plt.ylabel('Contagem semanal de bicicletas')
plt.show()
```

```python
daily = data.resample('D').sum()
daily.rolling(30, center = True).sum().plot(style = [':', '--', '-'])
plt.ylabel('Contagem mÃ©dia por hora')
plt.show()
```

```python
daily.rolling(50, center=True,
              win_type='gaussian').sum(std=10).plot(style=[':','--', '-'])
plt.show()
```

```python
import numpy as np
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks= hourly_ticks, style=[':', '--', '-'])
plt.ylabel("Traffic according to time")
plt.show()
```
